```python
from marl.audio import transforms as T
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as S
%matplotlib inline
```

```python
def extract_contours(fpath, framerate=172.3, bins_per_octave=42, n_octaves=8, freq_min=27.5,
                     pitch_cont=80, max_gap=100, amp_thresh=0.9, dev_thresh=0.9):
    """Extract pitch contours from an audio file.

    Parameters
    ----------
    fpath : string
        Path to audio file.
    framerate : float
        Number of CQT frames per second.
    bins_per_octave: int
        Number of constant Q bins per octave.
    n_octaves : int
        Number of octaves in CQT. (Determines maximum frequency)
    freq_min : float
        Minimum frequency in CQT
    pitch_cont : float
        Pitch continuity threshold in cents.
    max_gap : float
        Threshold (in miliseconds) for how many values can be taken from S-.
    amp_thresh : float
        Threshold on how big a peak must be relative to the maximum in its frame.
    dev_thresh : float
        The maximum number of standard deviations below the mean a peak can be to survive.

    Returns
    -------
    contours : list of np.ndarrays (size n_vals x 3)
        Extracted contours. Arrays have rows (time_idx, freq_idx, amplitude).
    cqt : np.ndarray
        CQT computed for contour extraction
    times : numpy array
        Array of time values corresponding to indices of the CQT
    freqs : numpy array
        Array of frequency values corresponding to indices of the CQT
    all_peaks_copy : numpy array
        Labeled peaks before contour extraction. (For plotting only)
    """    
    
    # compute sample rate necessary for specified cqt range
    samplerate = np.round(freq_min*(2.0**n_octaves)*2.5)
    
    print "Computing CQT..."
    # compute cqt
    times, freqs, cqt = T.cqt(audiofile, freq_min=27.5, octaves=n_octaves, 
                              samplerate=samplerate, bins_per_octave=bins_per_octave, 
                              framerate=framerate)
    cqt = np.abs(cqt[0])
    
    print "Removing contours..."
    
    # find coordinates of peaks in frequency for each time frame of the cqt
    peaks = S.argrelmax(cqt, axis=1)
    # store result in a single array
    all_peaks = np.ones((len(peaks[0]), 4)) # last axis is a marker for later
    all_peaks[:, 0] = peaks[0] # time index
    all_peaks[:, 1] = peaks[1] # frequency index
    all_peaks[:, 2] = cqt[peaks[0], peaks[1]] # cqt magnitude
    
    # for each frame, mark peaks with amplitude below a threshold 
    # of the maximum peak amplitude in the frame with label -1
    unique_frames = np.unique(all_peaks[:,0])
    for frame in unique_frames:
        this_frame = all_peaks[:, 0] == frame
        max_amp = np.max(all_peaks[this_frame, 2])
        bad_peaks = all_peaks[:, 2] < amp_thresh*max_amp
        all_peaks[this_frame & bad_peaks, 3] = -1.0
    
    # find indices of surviving peaks
    surviving_peaks = all_peaks[:, 3] == 1.0
    # compute mean and standard deviation of amplitudes of survivors
    mean_peak = np.mean(all_peaks[surviving_peaks, 2])
    std_peak = np.std(all_peaks[surviving_peaks, 2])
    
    # mark peaks with amplitude too far below the mean with label -2
    all_peaks[all_peaks[:,2] < (mean_peak-(dev_thresh*std_peak)), 3] = -2.0
    
    all_peaks_copy = all_peaks.copy()
    
    print "Selecting contours..."
    
    cpb = 100.0*12.0/float(bins_per_octave) # number of cents per bin
    n_bin = int(round(float(pitch_cont)/cpb)) # num bins within pitch continuity thresh
    n_gap = max_gap*0.001*framerate # num time frames within maximum gap
    
    contours = [] # initialize contours
    
    # loop until there are no remaining peaks labeled with 1
    while len(all_peaks[all_peaks[:,3] == 1.0, :]) > 0:
        
        # initialize list of indices to be placed in this contour
        this_contour_idx = []
        
        # find largest peak labeled with 1
        max_peak = np.argmax(all_peaks[all_peaks[:, 3] == 1.0, 2])
        max_peak_idx = np.arange(len(all_peaks))[all_peaks[:, 3] == 1.0][max_peak]
        
        # time frame and freqency index of largest peak
        time_frame = all_peaks[max_peak_idx, 0]
        f0_idx = all_peaks[max_peak_idx, 1]
        
        # add max peak to contour and change label to 0
        this_contour_idx.append(max_peak_idx)
        all_peaks[max_peak_idx, 3] = 0
        
        # choose remaining peaks for this contour
        gap = 0
        while gap < n_gap:
            
            # go to next time frame
            time_frame = time_frame + 1
            
            # initialize candidates for next peak
            cands = np.arange(len(all_peaks))
            
            # restrict candidates to time frame
            cands = cands[all_peaks[:, 0] == time_frame]
            
            # split between good candidates S+ and bad ones S-
            bad_cands = cands[all_peaks[cands, 3] < 0]
            cands = cands[all_peaks[cands, 3] == 1.0]
    
            # restrict to frames that satisfy pitch continuity
            cands = cands[np.abs(all_peaks[cands, 1] - f0_idx) < n_bin]
            bad_cands = bad_cands[np.abs(all_peaks[bad_cands, 1] - f0_idx) < n_bin]
            
            # if there are no candidates, finish the contour
            if len(cands) == 0 and len(bad_cands) == 0:
                break
            # if there are no candidates in S+, use a candidate from S-
            # but increase the gap
            elif len(cands) == 0:
                cands = bad_cands
                gap = gap + 1
            # otherwise, we have a candidate in S+, so reset the gap to 0
            else:
                gap = 0
            
            # find candidate that is closest in frequency to target
            min_dist = np.argmin(np.abs(all_peaks[cands, 1] - f0_idx))
            min_idx = np.arange(len(all_peaks))[cands][min_dist]
    
            # add this peak to the contour, change its label to 0
            this_contour_idx.append(min_idx)
            all_peaks[min_idx, 3] = 0
            # update target frequency
            f0_idx = all_peaks[min_idx, 1]
        
        # add completed contour to contour list
        contours.append(all_peaks[this_contour_idx, 0:3])
    
    return contours, cqt, times, freqs, all_peaks_copy
```

```python
def plot_contours(contours, cqt):
    fig = plt.figure(figsize=(12,6))
    plt.imshow(cqt.T, origin='lower')
    for contour in contours:
        plt.plot(contour.T[0], contour.T[1], '.')
    plt.axis('tight')
    plt.show()
```

```python
def plot_peaks(all_peaks_copy):
    apc = all_peaks_copy
    good_peaks = apc[:, 3] == 1.0
    max_filt_peaks = apc[:, 3] == -1.0
    std_filt_peaks = apc[:, 3] == -2.0
    fig = plt.figure(figsize=(12,6))
    plt.plot(apc[:, 0], apc[:, 1], 'b.', ms=1, label='Peaks in S-')
    plt.plot(apc[good_peaks, 0], apc[good_peaks, 1], 'r.', ms=1, label = 'Peaks in S+')
    plt.axis('tight')
    plt.legend()
    plt.show()
```
