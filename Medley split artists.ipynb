```python
import os
import glob
import yaml
import re
```

```python
def read_file(f):
    
    with open(f, 'r') as fd:
        return ''.join(line for line in fd)
```

```python
metadata = sorted(glob.glob('/media/bmcfee/31892bae-2481-425a-836f-5d503baa3bdb/MedleyDB/Audio/*/*.yaml'))
```

```python
artists = {}
for mf in metadata:
    root = os.path.splitext(os.path.basename(mf))[0]
    
    m1 = yaml.load_all(read_file(mf))
    
    artists[root] = list(m1)[0]['artist'].replace(' Multitracks', '')
```
